# K-Means Clustering

K-means clustering is an iterative clustering algorithm that aims to find local maxima in each iteration.

Importing modules and KMeans classifier

```python
# Importing Modules
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
iris_df
```

Randomly assuming number of clusters= 5

```python
# Declaring Model
model = KMeans(n_clusters=5)

# Fitting Model
model.fit(iris_df.data)

# Predicitng a single input
predicted_label = model.predict([[7.2, 3.5, 0.8, 1.6]])

# Printing Predictions
print('predicted label',predicted_label)
```

```python
# Prediction on the entire data
#all_predictions = model.predict(iris_df.data)

y_pred=model.fit_predict(iris_df.data)
y_pred
```

The centers of each clusters 

```python
model.cluster_centers_ 
```

# Elbow Method to Find out the Optimum number of clusters

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(iris_df.data)
    kmeans.fit(iris_df.data)
    Error.append(kmeans.inertia_)    #inertia means within cluster distance.. Kmeans try to minimize the inertia
plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

Now the best n value is between 2 and 4 so let's implement KMeans using n=3

```python
kmeans_3= KMeans(n_clusters=3)
kmeans=kmeans_3.fit(iris_df.data)
y_pred_3= kmeans.fit_predict(iris_df.data)
y_pred_3
```

```python
kmeans.cluster_centers_
```
